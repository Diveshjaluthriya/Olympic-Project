## Q2 Now using the same procedure for coaches let's find out disciplines in which max no. of coaches are involved and same for the countries

```python
noc1 = coaches_data.NOC.unique()
coach_countries = coaches_data['NOC']
values1 = []
for i in noc1:
    count = 0
    for j in coach_countries:
        if(i == j):
            count = count +1
    values1.append(count)
df1 = pd.DataFrame(data = zip(noc1,values1),
                 columns =[ 'country', 'no_of_coaches'])

discipline = coaches_data.Discipline.unique()
coach_discipline = coaches_data['Discipline']
values2 = []
for i in discipline:
    count = 0
    for j in coach_discipline:
        if(i == j):
            count = count +1
    values2.append(count)
df2 = pd.DataFrame(data = zip(discipline,values2),
                 columns =[ 'discipline', 'no_of_coaches'])
```

Q1 What is the total no. of medals?

```python
print(medals_data['Bronze'].sum(),": Bronze")
print(medals_data['Silver'].sum(),": Silver")
print(medals_data['Gold'].sum(),": Gold")
```

That's something unpalatable! How can the no. of various medals be different ?
It can be seen from the data that the ranking system of ranking by total is based upon the total no. of medals secured.

I don't like both methods of ranking. Let us `devise` a method based on `weighted average` for the ranking where,
Gold, silver and bronze gives 3, 2 and 1 points respectively.

```python
data = medals_data[['Gold', 'Silver', 'Bronze']]
data
```

```python
weights = [3, 2, 1]
weighted_avg = data*weights
## print(weighted_avg)
total_weight = weighted_avg['Gold']+weighted_avg['Silver']+weighted_avg['Bronze']
## print(total_weight)
```

Uncomment the lines to understand what happened

```python
medals_data['Total_point'] = total_weight
with pd.option_context('display.max_rows', 100):
    display(medals_data)
```

**Now let us sort according to total_point.**
The difference between sorting in accordance with Rank and total_point is discernible in example of Japan,ROC and Britain
Britain secured 65 medals and has Total_point 130 as compared to Japans 58 medal and 126 Total_point. But according to conventional ranking system since japan secured more gold medals it is held above britain which seems a bit unfair.
Same for ROC and various other countries.

Difference between ranking in accordance with Rank by Total and total_point can be seen below

```python
with pd.option_context('display.max_rows', 100):
    display(medals_data.sort_values('Total_point', ascending = False))
```
